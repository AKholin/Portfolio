```
data = df.copy()
print(data.shape)
```

```
data.drop(labels=['CarName', 'doornumber', 'peakrpm', 'compressionratio', 'stroke', 'carheight'], axis=1, inplace=True)
print(data.shape)
data_list = list(data.columns)
print(data_list)
```

```
sig_cat_col = ['symboling','fueltype','aspiration', 'enginelocation', 'carbody','drivewheel','enginetype','cylindernumber', 'fuelsystem']
sig_num_col = ['wheelbase', 'enginesize', 'price', 'carlength', 'citympg', 'carwidth', 'boreratio', 'highwaympg', 'curbweight', 'horsepower']
print(sig_num_col)
```

```
dummies = pd.get_dummies(data[sig_cat_col])
print(dummies.shape)
dummies.head()
```

```
data = pd.concat([data, dummies], axis = 1)
print(data.shape)
data.head()
```

```
data.drop(sig_cat_col, axis = 1, inplace = True)
print(data.shape)
data.head()
```

```
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(data, train_size = 0.7, test_size = 0.3, random_state = 42)
```

```
train_mean = df_train[sig_num_col].mean(axis=0)
train_std = df_train[sig_num_col].std(axis=0)
print(train_mean.shape)
print(train_std.shape)
print(train_mean)
print(train_std)
df_train[sig_num_col] = (df_train[sig_num_col] - train_mean) / train_std
df_train.head()
```

```
df_test[sig_num_col] = (df_test[sig_num_col] - train_mean) / train_std
df_test.head()
```

```
corr_matrix = df_train.corr()
corr_matrix["price"].sort_values(ascending=False)
```

```
corr_price = list(corr_matrix['price'])
train_col = list(df_train.columns)
cor = list(zip(train_col, corr_price))
cor.sort(key=lambda para: para[0], reverse=False)
cor
```

```
y_train = df_train.pop('price')
x_train = df_train

y_test = df_test.pop('price')
x_test = df_test

print(y_train.shape)
print(x_train.shape)
print(y_test.shape)
print(x_test.shape)
```

```
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE

lm = LinearRegression()
lm.fit(x_train, y_train)

rfe = RFE(lm, 15)             
rfe = rfe.fit(x_train, y_train)
```

```
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```
min_price = df['price'].min()
max_price = df['price'].max()
price_range = max_price-min_price
print(min_price)
print(max_price)
print(price_range)
```

```
train_std['price']
```

```
%%time

# Linear

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)

lin_reg_pred = lin_reg.predict(x_test)
lin_reg_mae = mean_absolute_error(y_test, lin_reg_pred)
print('Lin reg MAE:', lin_reg_mae)
print('R2 score:', r2_score(y_test, lin_reg_pred))
print('MAE = {0:.2f} $, erorr = {1:.02f} %'.format(lin_reg_mae*train_std['price'], lin_reg_mae*train_std['price']*100/price_range))
```

```
%%time
from sklearn import linear_model

# Ridge

model = linear_model.Ridge(alpha=0.1)
model.fit(x_train, y_train)
model_pred = model.predict(x_test)
model_mae = mean_absolute_error(y_test, model_pred)
print('Ridge MAE:', model_mae)
print('R2 score:', r2_score(y_test, lin_reg_pred))
print('MAE = {0:.2f} $, erorr = {1:.02f} %'.format(model_mae*train_std['price'], model_mae*train_std['price']*100/price_range))
```

```
%%time
from sklearn import linear_model

# Lasso

model = linear_model.Lasso(alpha=0.1)
model.fit(x_train, y_train)
model_pred = model.predict(x_test)
model_mae = mean_absolute_error(y_test, model_pred)
print('MAE:', model_mae)
print('R2 score:', r2_score(y_test, lin_reg_pred))
print('MAE = {0:.2f} $, erorr = {1:.02f} %'.format(model_mae*train_std['price'], model_mae*train_std['price']*100/price_range))
```
